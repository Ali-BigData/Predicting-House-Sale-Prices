### Selection of Numeric features to know which features drive the sale price?

```python
# take only numerical columns
df_nummeric = data.select_dtypes(include=["float","int"])
cormat = df_nummeric.corr()

# comparing with target coloum.
sorted_corrs  = cormat['SalePrice'].abs()

import seaborn as sns
## Let's only keep columns with a correlation coefficient of larger than 0.4 (arbitrary, worth experimenting later!)

strong_corre = sorted_corrs[sorted_corrs > 0.4]
corrmat = cormat[strong_corre.index].corr()
sns.heatmap(corrmat)
plt.show()
```

<b> Highly SalePrice-Predictive Features


```python

## Drop columns with less than 0.4 correlation with SalePrice
transform_df = data.drop(sorted_corrs[sorted_corrs < 0.4].index, axis=1) # threshold = 0.4
transform_df
```

- Which columns are currently numerical but need to be encoded as categorical instead (because the numbers don't have any semantic meaning)?
- If a categorical column has hundreds of unique values (or categories), should we keep it? When we dummy code this column, hundreds of columns will need to be added back to the data frame.

```python
def select_features(df, coeff_threshold=0.4, uniq_threshold=8):
    numerical_df = df.select_dtypes(include=['int', 'float'])
    abs_corr_coeffs = numerical_df.corr()['SalePrice'].abs().sort_values()
    
    # Drop columns with less than 0.4 correlation with SalePrice
    df = df.drop(abs_corr_coeffs[abs_corr_coeffs < coeff_threshold].index, axis=1)
    
    
 # Create a list of column names from documentation that are *meant* to be categorical   
    nominal_features = df.select_dtypes(include='object').columns
    
    
    transform_cat_cols = []
    for col in nominal_features:
        if col in df.columns:
            transform_cat_cols.append(col)
    
    
    # Accounting How many unique values in each categorical column
    uniqueness_counts = df[transform_cat_cols].apply(lambda col: len(col.value_counts())).sort_values()
    
    # removing category columns with more that 8 uniques.or Aribtrary cutoff of 8 unique values (worth experimenting)

    drop_nonuniq_cols = uniqueness_counts[uniqueness_counts > 8].index
    df = df.drop(drop_nonuniq_cols, axis=1)
    #    MULTICLASSES 
   # Select just the remaining text columns and convert to categorical 
    text_cols = df.select_dtypes(include=['object'])
    for col in text_cols:
        df[col] = df[col].astype('category')
        
    # Create dummy columns and add back to the dataframe!
    
    df = pd.concat([df, pd.get_dummies(df.select_dtypes(include=['category']))], axis=1).drop(text_cols,axis=1)
    
    return df
```

### 5- Using Kfold in LinearRegression.

```python
def train_and_test(df,k):
    
    
    numeric_df = df.select_dtypes(include=['integer', 'float'])   
    
    lr = linear_model.LinearRegression()
    
    if k == 0:
        Y = numeric_df["SalePrice"] #  Select target
        X = numeric_df.drop("SalePrice", axis=1) #  Select features
   
        
        X_train, X_test, y_train, y_test = train_test_split(X,Y, random_state=1)
       

        lr.fit(X_train, y_train)
        predictions = lr.predict(X_test)
        mse = mean_squared_error(y_test, predictions)
        rmse = np.sqrt(mse)
        
        print("Training set score: {:.3f}".format(lr.score(X_train, y_train)))
        
#         rmse_compare['LinearRegression_Kfold=0'] = rmse
#         scores_compare['LinearRegression_Kfold=0'] = lr.score(X_test, y_test)

        return rmse
    
    if k == 1:
        lr = linear_model.LinearRegression()
        # Randomize *all* rows (frac=1) from `df` and return
        shuffled_df = df.sample(frac=1, )
        Y = shuffled_df["SalePrice"] #  Select target
        X = shuffled_df.drop("SalePrice", axis=1) #  Select features
   
       
        X_train, X_test, y_train, y_test = train_test_split(X,Y, random_state=1)
        
        lr.fit(X_train, y_train)
        predictions_one = lr.predict(X_test)        
        
        mse_one = mean_squared_error(y_test, predictions_one)
        rmse_one = np.sqrt(mse_one)
        
        lr.fit(X_test, y_test)
        predictions_two = lr.predict(X_train)        
       
        mse_two = mean_squared_error(y_train, predictions_two)
        rmse_two = np.sqrt(mse_two)
        
        avg_rmse = np.mean([rmse_one, rmse_two])
        print("Training set score: {:.3f}".format(lr.score(X_train, y_train)))
        print("Test set score: {:.3f}".format(lr.score(X_test, y_test)))
        print("rmse_average: {}".format(avg_rmse))
        
        rmse_compare['LinearRegression_Kfold=1'] = avg_rmse
        scores_compare['LinearRegression_Kfold=1'] = lr.score(X_test, y_test)
        
    else:
        numeric_df = df.select_dtypes(include=['integer', 'float'])
        features = numeric_df.columns.drop("SalePrice")
        kf = KFold(n_splits=k, shuffle=True)
        rmse_values = []
        train_values =[]
        test_values =[]
        for train_index, test_index, in kf.split(df):
            train = df.iloc[train_index]
            test = df.iloc[test_index]
            lr.fit(train[features], train["SalePrice"])
            predictions = lr.predict(test[features])
            mse = mean_squared_error(test["SalePrice"], predictions)
            rmse = np.sqrt(mse)
          
        
            rmse_values.append(rmse)
            train_values.append(lr.score(train[features], train["SalePrice"]))
            test_values.append(lr.score(test[features], test["SalePrice"]))
            
            
            
        
           
            

        train_avg = np.mean(train_values)
        print("Training set score: {:.3f}".format(train_avg))
        
        avg_test = np.mean(test_values)
        print("Training set score: {:.3f}".format(avg_test))
        
        avg_rmse = np.mean(rmse_values)
        print("rmse score : {:.3f}".format(avg_rmse))
        
     
        plt.scatter(rmse_values,test_values)
    
        plt.xlabel('RMSE_values')
        plt.ylabel('Testing score')
      
        rmse_compare['LinearRegression_Kfold=10'] = avg_rmse
        scores_compare['LinearRegression_Kfold=10'] = np.mean(test_values)
        return avg_rmse
    
    
    
    
       
        
```
